## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
df.head()
```

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
## add another column have to tell the equation to get it from the data frame https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html
subtotal= df["qty"]*df["unit_price"]
subtotal
#now need to add it to end of 
#tells us it is a <class 'pandas.core.series.Series'> 
# i want to create another subtotal called subtotal 
df["subtotal"]= subtotal
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
    #   thoughts -we have unit_weight and qty so do math like I am using a calculator-- 
    #   guessed this at first #shipping_over50lbs= df["unit_weight"]*df[unit_qty]>50= 7+df[["unit_weight"]*df[unit_qty]]
total_weight= df["unit_weight"]*df["qty"]

df["shipping price"]=total_weight.apply(lambda x: x*7 if x >50 else x*10)

df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
## at first thought to muliply by 0.0925 but should be 1.0925
total_price = round((df["subtotal"]+df["shipping price"])*[(1.0925)],2)
total_price.head()
df["total price"]= total_price
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

total_cost= (df["unit_cost"]*df["qty"])+ df["shipping price"]
total_cost.head()
df["total cost"]= total_cost
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
## so total cost is line cost and total price is line price

profit= df["total price"]- df["total cost"]
profit
df["profit"]= profit
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
df.head()
```

```python
# How much did each of the top 5 clients by quantity spend? 
    # Check your work from Part 1 for client ids. 
    # need to group by top clients then count sum of total spent 
#get unique ids then get total sum by each clients look at 5 who spent the most then create df of top clients 
    #called it index_list we know that is top 5 - how much intotal each of 5 we are creating a 
    # function called top client and say which inputs we need for function to run that is why we use def 
    #total_price
    #def top_client(client_id,total_price): defined the functions and the inputs it is taking

def top_client(client_id,total_price):
    client_df = df.loc[df['client_id'] == client_id, total_price]
    return round(client_df.sum(),2)

#f strings when we want to combine written word with variables as stored values {}
for client_id in index_list:
    print(f"{client_id}:${top_client(client_id, 'total price')}")
    
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
   
data_dict = {'client_id': index_list}
columns = ['qty', 'shipping price', 'total price', 'total cost', 'profit']

for col in columns:
    data_dict[col] = [top_client(client_id, col) for client_id in index_list]

final_df= pd.DataFrame(data_dict)
final_df
```
